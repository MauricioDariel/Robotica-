## Definicion de Nodo y Topicos en ROS 2

Es la unidad de ejecucion basica de un sistema robotico, como un programa independiente que se encarga de realizar una tarea especifica,gestionar hardware o implementar un algoritmo. Los nodos se comunican entre si de forma modular y descentralizada a traes de temas donde publican o suscriben mensajes y servicios, los nodos se caracterizan por que estos son modulares e independientes

## Topicos  en ROS 2

Un topico es un canal de comunicacion anonimo y fuertemente tipado donde los publicadores envian mensajes y los subcritpores los reciben.Se usan para flujos continuos de datos.Publicador y subscriptor deben usar el mismo npmbre de topico y el mismo tipo de mensaje para intercambia rinformacion

## Los tres tipos principales de interfaces en ROS 2

Mensajes: Se usan para el flujo continuo de datos, como la telemetría de un sensor (por ejemplo, datos de un LiDAR) o el estado de un robot (por ejemplo, su posición y velocidad).


Servicios: Ideal para tareas que se ejecutan rápidamente y que requieren una respuesta inmediata. Por ejemplo, consultar el estado de un nodo o realizar un cálculo rápido como cinemática inversa.


Acciones: Se utilizan para objetivos que tardan tiempo en completarse tareas largas y cancelables (como mover un robot hasta un objetivo).

## Documentar la creacion de paquetes Arbol

Para documentar la creación y la estructura de un paquete en ROS 2, es fundamental combinar varios elementos: una descripción clara en el archivo `package.xml`, comentarios en el código, una estructura de directorios consistente, un archivo `README.md` detallado y el uso de herramientas de documentación automatizada como `rosdoc2`.

#### En el archivo `package.xml`

Este archivo es la primera parada para cualquier persona que quiera entender tu paquete. Debe contener metadatos precisos.

-  **`<name>`**: Nombre del paquete.
-  **`<version>`**: Versión actual.
-  **`<description>`**: Una breve descripción del propósito del paquete. Debe ser clara y concisa.
-  **`<maintainer>`**: Nombre y correo electrónico del mantenedor.
-  **`<license>`**: Licencia de código abierto del paquete (por ejemplo, Apache 2.0).
-  **`<depend>`**: Dependencias de tiempo de compilación y ejecución.
#### Archivo `README.md`

Este archivo es la documentación más visible y debe estar en la raíz de tu paquete.

-  **Resumen del paquete**: Qué hace el paquete, por qué es útil y qué problema resuelve.
-  **Requisitos**: Versión de ROS 2 y dependencias que no se gestionan a través de `package.xml`.
-  **Estructura del directorio**: Un árbol de directorios que explique la función de cada carpeta principal (`src`, `include`, `launch`, `config`, etc.). Puedes generarlo fácilmente con un comando como `tree` en Linux.
-  **Instrucciones de compilación**: Cómo compilar el paquete con `colcon`.
-  **Instrucciones de uso**: Cómo ejecutar los nodos o lanzar el paquete.
-  **Nodos y parámetros**: Descripción de los nodos que contiene, los tópicos que publican/suscriben, los servicios y las acciones que ofrecen, y los parámetros que aceptan.
-  **Ejemplos**: Fragmentos de código o comandos para ilustrar su uso.
####  Comentarios en el código

Una buena documentación debe empezar dentro del propio código.

-  **Comentarios de archivo**: Al inicio de cada archivo (`.cpp` o `.py`), incluye una descripción de su propósito y autor.
-  **Comentarios de funciones y clases**: Explica la lógica, los argumentos de entrada, los valores de retorno y cualquier comportamiento particular.
-  **ROS 2 CLI**: Cuando uses la línea de comandos de ROS 2 para crear el paquete, las plantillas generadas ya incluyen comentarios básicos que te sirven de punto de partida.
#### Herramienta `rosdoc2`

`rosdoc2` es la herramienta estándar para generar documentación de ROS 2 de forma automatizada, utilizando `Sphinx` para Python y `Doxygen` para C++.
